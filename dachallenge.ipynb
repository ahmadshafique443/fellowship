## Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading dataset

```python
tips = pd.read_csv('tips.csv')
```

```python
tips.head(5)
```

## 1) Delete the Unnamed 0 column

```python
tips.drop(columns = 'Unnamed: 0', inplace=True)
tips.head(5)
```

## 2) Plot the total_bill column histogram

```python
plt.figure(figsize=(7,5))
plt.hist(tips.total_bill, color='#f7a9a9')
plt.title('Total Bill Amount Distribution', fontdict={'fontweight':'bold','fontsize':16})
plt.xlabel('Total Bill Amount')
plt.ylabel('Count')
plt.show()
```

## 3) Create a scatter plot presenting the relationship between total_bill and tip

```python
plt.figure(figsize=(8,4))
plt.scatter(x='total_bill', y='tip', data=tips, color='#1ec405', alpha=0.5)
plt.title('Relationship Between Total Bill and Tip', fontdict={'fontweight':'bold','fontsize':18})
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

## 4) Create one image with the relationship of total_bill, tip and size

```python
sns.relplot(x='total_bill', y='tip', kind='scatter', hue='size', size='size', data=tips, height=4.4, aspect=1.7)
plt.title('Relationship Between Total Bill,Tip and Size', fontdict={'fontweight':'bold','fontsize':18})
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

## 5) Present the relationship between days and total_bill value

```python
day_bill = pd.DataFrame(tips.groupby('day')['total_bill'].sum())
day_bill.reset_index(inplace=True)
day_bill
```

```python
plt.figure(figsize=(7,5))
plt.bar(x='day', height='total_bill', data=day_bill, color='#96648b')
plt.title('Relationship Between Days and Total Bill Amount', fontdict={'fontweight':'bold','fontsize':16})
plt.xlabel('day')
plt.ylabel('Total Bill Amount')
plt.show()
```

## 6) Create a scatter plot with the day on y-axis and tip on x-axis, differ the dots by sex

```python
sns.relplot(x='tip', y='day', data=tips, size='sex', kind='scatter', height=4.3, aspect=1.7)
plt.title('Relationship Between Tip and Day', fontdict={'fontweight':'bold','fontsize':20})
plt.xlabel('Tip')
plt.ylabel('Day')
plt.show()
```

## 7) Create a box plot presenting the total_bill per day differentiation the time (Dinner or Lunch)

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='day', y='total_bill', data=tips, hue='time')
plt.title('Total Bill per Day Differentiation by Time', fontdict={'fontweight':'bold','fontsize':20})
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

## 8) Create two histograms of the tip value based for Dinner and Lunch side by side.

```python
sns.displot(x='tip', data=tips, kind='hist', hue='time', col='time')
plt.show()
```

## 9) Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker', col='sex')
plt.show()
```
