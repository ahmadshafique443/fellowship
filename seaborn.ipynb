# SEABORN

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Scatter plot

```python
student = pd.read_csv('student-alcohol-consumption.csv')
student.head()
```

```python
sns.relplot(x="absences", y="G3", data=student, kind='scatter',size='study_time',hue='study_time', height=5, aspect=1,  hue_order=['<2 hours', '2 to 5 hours','5 to 10 hours', '>10 hours'])
plt.title('Absence vs Grades relation')
plt.legend(loc='upper right')
plt.show()
```

### does a student's first semester grade "G1" tend to correlate with their final grade "G3"

```python
sns.relplot(x="G1", y="G3", 
            data=student,
            kind="scatter", 
            col="schoolsup",
            row='famsup',
            col_order=["yes", "no"],
           row_order=['yes','no'], height=2.5, aspect=1.3)
plt.show()
```

```python
mpg = pd.read_csv('mpg.csv')
mpg.head()
```

### Count Plot

```python
palette_colors = {"Rural": "#fa6616", 'Urban': "#fe7c73"}
plt.figure(figsize=(5,4))
# Create a count plot of school with location subgroups
sns.countplot(x='school', data=student, hue='location', palette=palette_colors)
plt.title('Students residence', fontdict={'fontweight':'bold','fontsize':18})


# Display plot
plt.show()
```

```python
survey = pd.read_csv('young-people-survey-responses.csv')
survey.head()
```

```python
sns.countplot(x='Internet usage', data=survey)
plt.xticks(rotation=10)
plt.show()
```

```python
category_order = ["<2 hours", "2 to 5 hours", "5 to 10 hours", ">10 hours"]
sns.catplot(x="study_time", y="G3", data=student, kind="bar", order=category_order, errorbar=None, height=3.5, aspect=1.2)
plt.title('Study time VS Grades', fontdict={'fontweight':'bold','fontsize':18})
plt.xticks(rotation=10)
plt.show()
```

```python
sns.relplot(x='horsepower', y='mpg', data=mpg, kind='scatter', hue='cylinders', size='cylinders', height = 4.4, aspect=1.7)
plt.title('  horsepower and cylinders impact on mpg', fontdict={'fontweight':'bold','fontsize':16})
plt.show()
```

```python
sns.relplot(x="model_year", y="horsepower", data=mpg, kind="line", errorbar=None, style="origin", 
            hue="origin", dashes=False, markers=True, height=4.4, aspect=1.7)
plt.title('  Average Horsepower for Cars over the Time', fontdict={'fontweight':'bold','fontsize':16})
plt.show()
```

### Point Plot

```python
sns.catplot(x="romantic", y="absences", data=student, kind="point", hue="school", errorbar=None, estimator=np.median)
plt.show()
```

### HeatMap

```python
show_guests = pd.read_csv('daily_show_guests_cleaned.csv')
show_guests.head()
```

```python
pd_crosstab = pd.crosstab(show_guests["Group"], show_guests["YEAR"])
plt.figure(figsize=(9, 6))
sns.heatmap(pd_crosstab)
plt.xticks(rotation=25)
plt.show()
```

```python
flights = sns.load_dataset('flights')
custom_palette = sns.color_palette( 'pastel', desat=.8)
plt.figure(figsize=(8,5))
sns.heatmap(data = flights.pivot('month', 'year', 'passengers'), annot = flights.pivot('month', 'year', 'passengers'), 
            fmt = 'd', linewidths = .4, cmap = custom_palette)
plt.xticks(rotation = 10)
plt.yticks(rotation = 0)
plt.title('Total Passengers in each month', fontdict={'fontweight':'bold','fontsize':18})
plt.show()
```

### Stacked Histogram

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
plt.figure(figsize=(4.3,5.5))
g= sns.histplot(tips, x="total_bill", hue='sex', multiple='stack')
plt.title('Total Bill VS Gender', fontdict={'fontweight':'bold','fontsize':16})
plt.show()
```
