# Matplotlib

#### importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
```

#### loading 'gas_prices' dataset

```python
gas_prices = pd.read_csv('gas_prices.csv')
gas_prices.head()
```

#### Line plot

```python
plt.figure(figsize=(10,5))
#plt.plot(gas_prices.Year, gas_prices.USA, label = 'USA')
#plt.plot(gas_prices.Year, gas_prices.Australia, label = 'Australia')
#plt.plot(gas_prices.Year, gas_prices.France, label = 'France')
#plt.plot(gas_prices.Year, gas_prices.Italy, label = 'Italy')
for country in gas_prices:
    if country != 'Year':
        plt.plot(gas_prices.Year, gas_prices[country], label = country, marker='.')

plt.xticks(gas_prices.Year[::2])
plt.title('Gas Price 1990-2008', fontdict ={'fontweight':'bold','fontsize':20})
plt.xlabel('Year', fontsize=14)
plt.ylabel('Price in USD', fontsize = 14)
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
```

##### Avg price in a country 1990-2008

```python
df = []
for i in range(1,11):
    df.append(round(gas_prices.iloc[:,i].mean(),2))
countries = gas_prices.columns[1:]
avg_price = pd.DataFrame({'Country' : countries,'Avg_gas_price' : df})
avg_price
```

#### Bar plot

```python
plt.figure(figsize=(8,4))
colors = ['red', 'blue', 'green', 'yellow', 'cyan', 'magenta', 'orange', 'purple', 'brown', 'pink']
plt.bar(avg_price['Country'], avg_price['Avg_gas_price'], color = colors)
plt.xticks(rotation = 20)
plt.yticks(np.arange(0,5.5,.5))
plt.title('Avg gas price in each country (1990-2008)', fontdict ={'fontweight':'bold','fontsize':20})
plt.xlabel('Country', fontsize=14)
plt.ylabel('Price in USD', fontsize = 14)
plt.show()
```

#### Loading 'fifa_data' dataset

```python
fifa = pd.read_csv('fifa_data.csv',index_col='index')
fifa.head(5)
```

#### Histogram

```python
plt.figure(figsize=(8,5))
bin = np.arange(16,47,3)
plt.hist(fifa.Age, bins = bin, color = '#b570a3')
plt.xticks(bin)
plt.title('Players Age Distriution', fontdict ={'fontweight':'bold','fontsize':20})
plt.xlabel('Age', fontsize=14)
plt.ylabel('Total players', fontsize = 14)
plt.show()
```

### Pie Chart

```python
right = fifa[fifa['Preferred Foot']=='Right'].count()[0]
left = fifa[fifa['Preferred Foot']=='Left'].count()[0]
```

```python
plt.figure(figsize=(7,5))
plt.pie([left, right], labels = ['Left', 'Right'], autopct='%.2f%%')
plt.title('Players Foot preference', fontdict ={'fontweight':'bold','fontsize':18})
plt.show()
```

```python
def str_to_num(x):
        if isinstance(x, float):
            return x
        elif 'K' in x:
            return float(x.strip('€K')) * 1000
        elif 'M' in x:
            return float(x.strip('€M')) * 1000000
        else:
            return float(x)
```

```python
fifa['Release Clause'] = fifa['Release Clause'].apply(lambda x:str_to_num(x))
p_above_50M = fifa[fifa['Release Clause']>=50000000].count()[0]
p10M_to_50M = fifa[(fifa['Release Clause']<50000000) & (fifa['Release Clause']>=10000000)].count()[0]
p5M_to_10M = fifa[(fifa['Release Clause']< 10000000) & (fifa['Release Clause']>=5000000)].count()[0]
p1M_to_5M = fifa[(fifa['Release Clause']<  5000000) & (fifa['Release Clause']>= 1000000)].count()[0]
p500K_to_1M = fifa[(fifa['Release Clause']<1000000) & (fifa['Release Clause']>= 500000)].count()[0]
p100K_to_500K = fifa[(fifa['Release Clause']<500000) & (fifa['Release Clause']>=100000)].count()[0]
p_under100K = fifa[(fifa['Release Clause']< 10000000)].count()[0]
```

```python
plt.figure(figsize=(5,5))
player_count = [p_above_50M, p10M_to_50M, p5M_to_10M, p1M_to_5M, p500K_to_1M, p100K_to_500K, p_under100K]
labels = ['€50M+', '€10M-€50M', '€5M-€10M', '€1M-€5M', '€500K-€1M', '€100K-€500K', '€0-€100K']
explode = (.1,.1,.1,0,0,0,0)
plt.pie(player_count, labels=labels, explode=explode, autopct='%.2f%%')
plt.title('Players Count as per their Release Clause range', fontdict ={'fontweight':'bold','fontsize':18})
plt.show()
```

### Scatter Plot
